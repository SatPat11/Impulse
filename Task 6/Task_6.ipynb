```python
import pandas as pd
```

```python
file_path_complex="/content/complex_all_features_extra.csv"
file_path_electro="/content/electro_all_features_extra.csv"
file_path_normal="/content/normal_all_features_extra.csv"
file_path_video="/content/video_all_features_extra.csv"




file_path_complex_valid="/content/complex_all_features_extra.csv"
file_path_electro_valid="/content/valid/electro_valid_all_features_extra.csv"
file_path_normal_valid="/content/valid/normal_valid_all_features_extra .csv"
file_path_video_valid="/content/valid/video_valid_all_features_extra.csv"









```

```python
df_complex= pd.read_csv(file_path_complex)
df_electro= pd.read_csv(file_path_electro)
df_normal= pd.read_csv(file_path_normal)
df_video= pd.read_csv(file_path_video)



df_complex_valid= pd.read_csv(file_path_complex_valid)
df_electro_valid= pd.read_csv(file_path_electro_valid)
df_normal_valid= pd.read_csv(file_path_normal_valid)
df_video_valid= pd.read_csv(file_path_video_valid)


```

```python
# Specify the columns to remove
columns_to_remove = ['Median Frequency', 'Frequency IQR', 'frequency', 'file_name', 'channel']

# Remove the columns from each DataFrame
df_complex = df_complex.drop(columns=columns_to_remove)
df_electro = df_electro.drop(columns=columns_to_remove)
df_normal = df_normal.drop(columns=columns_to_remove)
df_video = df_video.drop(columns=columns_to_remove)



df_complex_valid = df_complex_valid.drop(columns=columns_to_remove)
df_electro_valid = df_electro_valid.drop(columns=columns_to_remove)
df_normal_valid = df_normal_valid.drop(columns=columns_to_remove)
df_video_valid = df_video_valid.drop(columns=columns_to_remove)
```

```python
df_complex
```

```python
df_complex['label'] = 1
df_electro['label'] = 2
df_video['label'] = 3
df_normal['label'] = 0


df_complex_valid['label'] = 1
df_electro_valid['label'] = 2
df_video_valid['label'] = 3
df_normal_valid['label'] = 0


```

```python
df_complex_valid
```

```python
# Concatenate the DataFrames
combined_df = pd.concat([df_complex, df_electro, df_video, df_normal], ignore_index=True)


combined_df_valid= pd.concat([df_complex_valid, df_electro_valid, df_video_valid, df_normal_valid], ignore_index=True)
```

```python
combined_df
```

```python
# Shuffle the rows
combined_df = combined_df.sample(frac=1).reset_index(drop=True)

combined_df_valid = combined_df_valid.sample(frac=1).reset_index(drop=True)
```

```python
combined_df_valid
```

```python
from sklearn.model_selection import train_test_split

# Define the features (column names) you want to use for training
features = combined_df.columns.tolist()  # Get all columns as features
features.remove('label')  # Remove the 'label' column as it's the target

# Split into training and remaining data (80% train, 20% remaining)
X_train, X_rem, y_train, y_rem = train_test_split(combined_df[features], combined_df['label'], test_size=0.2, random_state=42)

# Split remaining data into validation and test sets (10% each from original data)
X_valid, X_test, y_valid, y_test = train_test_split(X_rem, y_rem, test_size=0.5, random_state=42)

X_validation=combined_df_valid[features]
y_validation=combined_df_valid['label']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Get the feature names (column names excluding the label)
feature_names = X_train.columns

# Fit the scaler on the training data only
scaler.fit(X_train)

# Transform all datasets using the fitted scaler and convert back to DataFrames
X_train = pd.DataFrame(scaler.transform(X_train), columns=feature_names, index=X_train.index)
X_rem = pd.DataFrame(scaler.transform(X_rem), columns=feature_names, index=X_rem.index)
X_valid = pd.DataFrame(scaler.transform(X_valid), columns=feature_names, index=X_valid.index)
X_test = pd.DataFrame(scaler.transform(X_test), columns=feature_names, index=X_test.index)






```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Create the SVM model
svm_model = SVC()

# Create GridSearchCV object
grid_search = GridSearchCV(svm_model, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to your data
grid_search.fit(X_train_subset, y_train_subset)

# Get the best model and its parameters
best_svm_model = grid_search.best_estimator_
best_params = grid_search.best_params_

# Evaluate the best model on your test data
# ...
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming you have:
# - best_svm_model: Your best SVM model obtained from hyperparameter tuning
# - X_test: Your test data
# - y_test: The true labels for the test data

# Make predictions on the test data
y_pred = best_svm_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Generate classification report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
